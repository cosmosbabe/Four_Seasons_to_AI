In math and science we encounter many situations in which analytical integration of a function or a differential equation is difficult or impossible. In this section I will show how Scientific Python can help through its high level mathematical algorithms. A simple trapezium formula calculates the integral of a function $f(x)$ as the area under the curve representing $f(x)$ by approximating it with the sum of trapeziums and the area of each trapezium is calculated as width times the average height.

$$\int_0^1 \frac{x^4(1 - x)^4}{1 + x^2} dx$$ 

The LaTeX formulas are beautiful, aren't they?! Let's write a small program to evaluate the integral shown above:

```python
def f(x):
    return x**4 * (1 - x)**4 / (1 + x**2)

#define a function to do integration of f(x) btw. 0 and 1:

def trapezium(f, n):
    h = 1 / float(n)
    integral = 0.5 * h * (f(0) + f(1))
    
    for i in range(1, int(n)):
        integral = integral + h * f(i * h)
    return integral

print(trapezium(f, 100))
```

Question: How many trapeziums are needed to get a relative error of less than 1 part in 1,000,000? Computation follows:

```python
n = 1
while abs(trapezium(f, n) - trapezium(f, n * 2)) > 1e-6:
    n += 1
    
print(n)
```

At this point we will modify the previous program because of the infinite range of integration and check the integrand to see where it becomes negligible. First, we pause and enjoy the beauty of the elegant integral :) Then we calculate:

$$ \int_{-\infty}^{\infty} e^{-x^2} dx $$

```python
def f(x):
    return math.exp(-x**2)

#define a function to do integration of f(x) btw. a and b:

def trapezium(f, n, a ,b):
    h = (b - a) / float(n)
    integral = 0.5 * h * (f(a) + f(b))
    
    for i in range(1, int(n)):
        integral = integral + h * f(a + i * h)
    return integral

a = -10
b = 10
n = 100

while (abs(trapezium(f, n, a, b) - trapezium(f, n*4, a*2, b*2)) > 1e-6):
    n *= 4
    a *= 2
    b *= 2
    
print(trapezium(f, n, a, b))
```

There is art in mathematics indeed. In 1963 Edward Lorenz, an American mathematician and meteorologist, published his famous set of coupled nonlinear first-order ordinary differential equations. They are relative simple but the resulting behavior is wonderfully complex. The equations are:

$ dx/dt = s(y - x) $

$ dy/dt = rx - y - xz $

$ dz/dt = xy - bz $

The solution executes a trajectory, plotted in three dimensions, that winds around and around, neither predictable nor random, occupying a region known as its $attractor$. Lorenz' work is important because he had an insight of the essence of chaos and found a unimodal structure in the flow of the solution. And as we know, order emerges from chaos.

```python
def lorenz(xyz, *, s = 10, r = 28, b = 3):
    """
    xyz: array-like, shape(3)
        Point of interest in three-dimensional space.
    s, r, b: int
        Parameters defining the Lorenz attractor.
    """
    
    x, y, z = xyz
    x_dot = s * (y - x)
    y_dot = r * x - y - x * z
    z_dot = x * y - b * z
    return np.array([x_dot, y_dot, z_dot])

dt = 0.01
num_steps = 1000

xyzs = np.empty((num_steps + 1, 3))   # Need one more for initial values
xyzs[0] = (0., 1., 1.05)   # Set initial values

# Step through time, calculating the partial derivatives at the current point
# and using them to estimate the next point

for i in range(num_steps):
    xyzs[i + 1] = xyzs[i] + lorenz(xyzs[i]) * dt
    
ax = plt.figure().add_subplot(projection = '3d')

ax.plot(*xyzs.T, lw = 1)
ax.set_xlabel("X Axis")
ax.set_ylabel("Y Axis")
ax.set_zlabel("Z Axis")
ax.set_title("Plot of the Lorenz attractor")

plt.show()
```
